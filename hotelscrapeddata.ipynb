```python
# Importing Libraries
from selenium.webdriver import Chrome
import pandas as pd
```

```python
# Using Chrome Driver to open a new window automatically
driver = Chrome(executable_path = "C:\\Users\\veni_\\Downloads\\New folder\\chromedriver-win64\\chromedriver.exe")
driver.maximize_window()
```

```python
driver.get("https://hotel.yatra.com/hotel-search/dom/search?checkoutDate=19/09/2023&checkinDate=17/09/2023&roomRequests[0].id=1&roomRequests[0].noOfAdults=2&roomRequests[0].noOfChildren=0&source=BOOKING_ENGINE&pg=1&tenant=B2C&isPersnldSrp=1&city.name=Jaipur&city.code=Jaipur&state.name=RJ&state.code=RJ&country.name=India&country.code=IND")
```

```python
# Code to extract data for name of the Hotel

hotels = []
for i in range(10):
    print("Scraping page", i + 1)
    hotel = driver.find_elements_by_xpath('//a[@class="under-link ng-binding"]')
    for h in hotel:
        hotels.append(h.text)
    
    
print(hotels)
```

```python
# Code to extract data for area and the city in which the hotel is situated

areas = []
for i in range(10):
    print("Scraping page", i + 1)
    area = driver.find_elements_by_xpath('//span[@class="va-sub ng-binding"]')
    for h in area:
        areas.append(h.text)
print(areas)
```

```python
# Code to extract data for prices of the various hotels in Jaipur

prices = []
for i in range(10):
    print("Scraping page", i + 1)    
    price = driver.find_elements_by_xpath('//span[@class="bold ng-binding"]')
    for h in price:
        prices.append(h.text)
print(prices)
```

```python
# Code to extract the ratings of various hotels in Jaipur

ratings = []
for i in range(10):
    print("Scraping page", i + 1)
    rating = driver.find_elements_by_xpath('//span[@class="b ng-binding"]')
    for h in rating:
        ratings.append(h.text)
print(ratings)
driver.close()
```

```python
# creating a dictionary

d = {"Hotel" : hotels, "Area": areas, "Ratings": ratings, "Price" : prices}
```

```python
# Convrerting the fetchced data into dataframe 

df = pd.DataFrame.from_dict(d, orient='index')
df = df.transpose()
df
```
